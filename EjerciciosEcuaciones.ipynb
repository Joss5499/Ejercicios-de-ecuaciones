Ejercicio 1

a) Resuelve por sustitución:

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 1 (a)
# Sistema:
# 5x + 2y = 1
# -3x + 3y = 5

A = np.array([[5, 2],
              [-3, 3]])
B = np.array([1, 5])

# Resolver sistema
sol = np.linalg.solve(A, B)
x, y = sol

print(f"Ejercicio 1 (a) - Solución:\nx = {x}\ny = {y}")

# Graficar sistema
x_vals = np.linspace(-10, 10, 400)

# Ecuaciones en forma y = ...
y1 = (1 - 5*x_vals) / 2
y2 = (5 + 3*x_vals) / 3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label='5x + 2y = 1')
plt.plot(x_vals, y2, label='-3x + 3y = 5')
plt.scatter(x, y, color='red', label='Solución')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 1 (a)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 6 (a)
A = np.array([[1, 2], [-3, 1]])
B = np.array([1, -10])

sol = np.linalg.solve(A, B)
x, y = sol

print("Ejercicio 6 (a)")
print(f"x = {x}, y = {y}")

x_vals = np.linspace(-10, 10, 400)
y1 = (1 - x_vals) / 2
y2 = (10 + 3 * x_vals)

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label='x + 2y = 1')
plt.plot(x_vals, y2, label='-3x + y = -10')
plt.scatter(x, y, color='red', label='Solución')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 6 (a)')
plt.legend()
plt.grid(True)
plt.show()
```

b)


```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz y vector
A = np.array([[-1, 2], [2, -4]])
B = np.array([4, 3])

# Solución aproximada por mínimos cuadrados
sol, residuals, rank, s = np.linalg.lstsq(A, B, rcond=None)
x, y = sol

print("Ejercicio 6 (b) - solución aproximada")
print(f"x = {x:.4f}, y = {y:.4f}")
print(f"Residuals (error cuadrático): {residuals}")

# Graficar las rectas
x_vals = np.linspace(-10, 10, 400)
y1 = (4 + x_vals) / 2
y2 = (3 - 2 * x_vals) / -4

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label='-x + 2y = 4')
plt.plot(x_vals, y2, label='2x - 4y = 3')
plt.scatter(x, y, color='red', label='Solución aproximada')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 6 (b)')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 15 (a)
A = np.array([[1, 1], [2, 3]])
B = np.array([6, 15])

sol = np.linalg.solve(A, B)
x, y = sol

print("Ejercicio 15 (a)")
print(f"x = {x}, y = {y}")

x_vals = np.linspace(-5, 10, 400)
y1 = 6 - x_vals
y2 = (15 - 2 * x_vals) / 3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label='x + y = 6')
plt.plot(x_vals, y2, label='2x + 3y = 15')
plt.scatter(x, y, color='red', label='Solución')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ejercicio 15 (a)')
plt.legend()
plt.grid(True)
plt.show()
```

**PROBLEMAS DE SISTEMAS DE ECUACIONES**

Problema nº 1.-
Calcula un número sabiendo que la suma de sus dos cifras es 10; y que, si invertimos el orden de dichas
cifras, el número obtenido es 36 unidades mayor que el inicial.

```python
import numpy as np

# Variables: x = decena, y = unidad
# Ecuaciones:
# 1) x + y = 10
# 2) 10*y + x = 10*x + y + 36  -> número invertido = número original + 36

A = np.array([[1, 1], [-9, 9]])
B = np.array([10, 36])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 1:\nDecena = {x}, Unidad = {y}")
print(f"Número original = {int(x*10 + y)}")
```

Problema nº 2.-
En un triángulo rectángulo, uno de sus ángulos agudos es 12 mayor que el otro. ¿Cuánto miden sus tres
ángulos?


```python
import numpy as np

# Variables: x = ángulo menor, y = ángulo mayor
# Ecuaciones:
# 1) y = x + 12
# 2) x + y = 90  (porque ángulos agudos en triángulo rectángulo suman 90)

A = np.array([[0, 1], [1, 1]])
B = np.array([12, 90])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 2:\nÁngulo menor = {x}°, Ángulo mayor = {y}°, Ángulo recto = 90°")
```

Problema nº 3.-
La distancia entre dos ciudades, A y B, es de 255 km. Un coche sale de A hacia B a una velocidad de 90
km/h. Al mismo tiempo, sale otro coche de B hacia A a una velocidad de 80 km/h. Suponiendo su velocidad
constante, calcula el tiempo que tardan en encontrarse, y la distancia que ha recorrido cada uno hasta el
momento del encuentro

```python
# Variables:
# t = tiempo en horas hasta que se encuentran
# x = distancia que recorre el coche que sale de A (90 km/h)
# y = distancia que recorre el coche que sale de B (80 km/h)

# Ecuaciones:
# 1) x + y = 255 (distancia total)
# 2) x/90 = y/80 = t (tiempo igual para ambos)

# Usamos variables x,y y resolveremos para t al final

A = np.array([[1, 1]])
B = np.array([255])

# La segunda ecuación es: x/90 = y/80 -> 80x = 90y -> 80x - 90y = 0

A = np.vstack([A, [80, -90]])
B = np.append(B, 0)

sol = np.linalg.lstsq(A, B, rcond=None)[0]
x, y = sol
t = x / 90

print(f"Problema 3:")
print(f"Tiempo hasta encuentro: {t:.2f} horas")
print(f"Distancia recorrida por coche A: {x:.2f} km")
print(f"Distancia recorrida por coche B: {y:.2f} km")
```

Problema nº 6.-
La razón entre las edades de dos personas es de 2/3. Sabiendo que se llevan 15 años, ¿cuál es la edad de cada
una de ellas?

```python
# Variables: x = edad de persona 1, y = edad de persona 2
# Ecuaciones:
# 1) x/y = 2/3 -> 3x - 2y = 0
# 2) |x - y| = 15

# Suponemos x < y
# 2) y - x = 15

A = np.array([[3, -2], [-1, 1]])
B = np.array([0, 15])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 6:\nEdad persona 1 = {x}, Edad persona 2 = {y}")
```

Problema nº 7.-
Un número excede en 12 unidades a otro; y si restáramos 4 unidades a cada uno de ellos, entonces el primero
sería igual al doble del segundo. Plantea un sistema y resuélvelo para hallar los dos números.

```python
# Variables: x = primer número, y = segundo número
# Ecuaciones:
# 1) x = y + 12
# 2) x - 4 = 2(y - 4)

A = np.array([[1, -1], [1, -2]])
B = np.array([12, -4])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 7:\nPrimer número = {x}, Segundo número = {y}")
```

Problema nº 8.-
El perímetro de un triángulo isósceles es de 19 cm. La longitud de cada uno de sus lados iguales excede en 2
cm al doble de la longitud del lado desigual. ¿Cuánto miden los lados del triángulo?

```python
# Variables: x = lado desigual, y = lados iguales
# Ecuaciones:
# 1) Perímetro: x + 2y = 19
# 2) y = 2x + 2

A = np.array([[1, 2], [-2, 1]])
B = np.array([19, 2])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 8:\nLado desigual = {x} cm, Lados iguales = {y} cm")
```

Problema nº 10.-
La suma de las tres cifras de un número capicúa es igual a 12. La cifra de las decenas excede en 4 unidades al
doble de la cifra de las centenas. Halla dicho número.

```python
# Variables: x = centenas, y = decenas, z = unidades
# Un número capicúa: x y x
# Ecuaciones:
# 1) x + y + x = 12 -> 2x + y = 12
# 2) y = 2x + 4

A = np.array([[2, 1], [-2, 1]])
B = np.array([12, 4])

sol = np.linalg.solve(A, B)
x, y = sol
z = x  # por ser capicúa

print(f"Problema 10:\nCifras: {int(x)}, {int(y)}, {int(z)}")
print(f"Número capicúa: {int(x*100 + y*10 + z)}")
```

Problema nº 11.-
El perímetro de un rectángulo es de 22 cm, y sabemos que su base es 5 cm más larga que su altura. Plantea un
sistema de ecuaciones y resuélvelo para hallar las dimensiones del rectángulo.

```python
# Variables: x = altura, y = base
# Ecuaciones:
# 1) 2(x + y) = 22 -> x + y = 11
# 2) y = x + 5

A = np.array([[1, 1], [-1, 1]])
B = np.array([11, 5])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 11:\nAltura = {x} cm, Base = {y} cm")
```

Problema nº 12.-
Hemos mezclado dos tipos de líquido; el primero de 0,94 €/litro, y el segundo, de
0,86 €/litro, obteniendo 40 litros de mezcla a 0,89 €/litro. ¿Cuántos litros hemos puesto de cada clase?

```python
# Variables: x = litros primer líquido, y = litros segundo líquido
# Ecuaciones:
# 1) x + y = 40
# 2) 0.94x + 0.86y = 0.89 * 40

A = np.array([[1, 1], [0.94, 0.86]])
B = np.array([40, 0.89 * 40])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 12:\nLitros primer líquido = {x:.2f}, Litros segundo líquido = {y:.2f}")
```

Problema nº 13.-
El doble de un número más la mitad de otro suman 7; y, si sumamos 7 al primero de ellos, obtenemos el

Problema nº 14.-
Dos de los ángulos de un triángulo suman 122. El tercero de sus ángulos excede en
4 grados al menor de los otros dos. ¿Cuánto miden los ángulos del triángulo?

```python
# Variables: x = ángulo menor, y = otro ángulo
# Ecuaciones:
# 1) x + y = 122
# 2) z = x + 4
# 3) x + y + z = 180

# De (3) y (2): x + y + x + 4 = 180 -> 2x + y = 176

A = np.array([[1, 1], [2, 1]])
B = np.array([122, 176])

sol = np.linalg.solve(A, B)
x, y = sol
z = x + 4

print(f"Problema 14:\nÁngulos: {x}°, {y}°, {z}°")
```

Problema nº 15.-
Una persona invierte en un producto una cantidad de dinero, obteniendo un 5% de beneficio. Por otra inversión
en un segundo producto, obtiene un beneficio del 3,5%. Sabiendo que en total invirtió 10 000 €, y que los
beneficios de la primera inversión superan en 300 € a los de la segunda, ¿cuánto dinero invirtió en cada
producto?

```python
# Variables: x = inversión 1, y = inversión 2
# Ecuaciones:
# 1) x + y = 10000
# 2) 0.05x = 0.035y + 300

A = np.array([[1, 1], [0.05, -0.035]])
B = np.array([10000, 300])

sol = np.linalg.solve(A, B)
x, y = sol

print(f"Problema 15:\nInversión 1 = {x:.2f} €, Inversión 2 = {y:.2f} €")
```
